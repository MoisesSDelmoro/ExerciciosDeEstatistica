```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

roteadores = ['Router1', 'Router2', 'Router3', 'Router4', 'Router5', 'Router6', 'Router7']

dataframe = pd.read_fwf('../Estatisticas'+'/data.txt',header=None,
names=["Router1", "Router2", "Router3", "Router4","Router5","Router6",
"Router7", "Room"])

def plot_boxplot(roteador):
    sala_i = dataframe.loc[:, [roteador, 'Room']]
    boxplot  = sala_i.boxplot(column=roteador, by='Room')
```

```python
for i in roteadores:
    plt.rcParams['figure.figsize'] = (7,5)
    plot_boxplot(i)    
```
